```python
import numpy as np
import matplotlib.pyplot as plt

#GLOBALES
k_theta = 0.033
def jexp(t_datos, k_theta):
  return (k_theta*t_datos)/(4*np.pi**2)

#MOMENTO INERCIA VARILLA

print('MOMENTO INERCIA VARILLA')
t_datos = np.array([1.90,1.81,1.91])**2
jexp0 = jexp(t_datos, k_theta)
jexp_medio = np.mean(jexp0)
for i in jexp0:
  print(f'El momento de inercia de una varilla es: {i:.4f} kgm2')

print('-----------------------------------------------------------')

print(f'El momento de inercia promedio de una varilla es: {jexp_medio:.4f} kgm2')

print('-----------------------------------------------------------')
mvar = 0.132
lvar = 0.6
jteo_varilla = (mvar*lvar**2)/12
print(f'El momento de inercia teórico de una varilla es: {jteo_varilla:.4f} kgm2')

print('-----------------------------------------------------------')

#MOMENTO INERCIA MASA + VARILLA (FALTA EL ULTIMO DATO)

momentos_masavarilla = np.array([])
print('MOMENTO INERCIA MASA + VARILLA')
t_datos1 = (np.array([9.94,12.72,15.85,20.44,22.9])/3)**2
jexp1= jexp(t_datos1, k_theta)
for i in jexp1:
  print(f'El momento de inercia de una masa + varilla es: {i:.4f} kgm2')
  momentos_masavarilla = np.append(momentos_masavarilla, i)

print('-----------------------------------------------------------')

#MOMENTO INERCIA MASA

momentos_masa = np.array([])
print('MOMENTO INERCIA MASA')
for i in jexp1:
    print(f'El momento de inercia de la masa es: {i-jexp_medio:.4f} kgm2')
    momentos_masa = np.append(momentos_masa, i-jexp_medio)

print('-----------------------------------------------------------')

#MOMENTO INERCIA CLINDRO MACIZO (112mm)

print('MOMENTO INERCIA CLINDRO MACIZO (112mm)')
t_datos2 = (np.array([5.49,5.34,5.36])/3)
jexp2 = jexp(t_datos2, k_theta)
jexp2_medio = np.mean(jexp2)
for i in jexp2:
  print(f'El momento de inercia del clindro macizo es: {i:.7f} kgm2')

print('-----------------------------------------------------------')

print(f'El momento de inercia promedio del cilindro macizo (112mm) es: {jexp2_medio:.7f} kgm2')

print('-----------------------------------------------------------')
m = 0.333
l = 0.112
jteo_clindro = (m*l**2)/2
print(f'El momento de inercia teórico del clindro macizo (112mm) es: {jteo_clindro:.7f} kgm2')

print('-----------------------------------------------------------')

#MOMENTO INERCIA DISCO SOPORTE

print('MOMENTO INERCIA CILINDRO SOPORTE')
t_datos3 = np.array([2.24,2.08,2.07])/3
jexp3 = jexp(t_datos3,k_theta)
jexp3_medio = np.mean(jexp3)

for i in jexp3:
  print(f'El momento de inercia del disco soporte es: {i:.7f} kgm2')
print('-----------------------------------------------------------')

print(f'El momento de inercia medio del disco soporte es: {jexp3_medio:.7f}')

print('-----------------------------------------------------------')

#MOMENTO INERCIA DISCO SOPORTE + CILINDRO MACIZO (44.6mm)

print('MOMENTO INERCIA CILINDRO SOPORTE + CILINDRO MACIZO')
t_datos4 = np.array([2.70,2.92,2.73])/3
jexp4 = jexp(t_datos4,k_theta)
jexp4_medio = np.mean(jexp4)

for i in jexp4:
  print(f'El momento de inercia del disco soporte + cilindro macizo es: {i:.7f} kgm2')
print('-----------------------------------------------------------')

print(f'El momento de inercia medio del disco soporte + cilindro hueco es: {jexp4_medio:.7f}')

print('-----------------------------------------------------------')

#MOMENTO INERCIA DISCO SOPORTE + CILINDRO HUECO

print('MOMENTO INERCIA CILINDRO SOPORTE + CILINDRO HUECO')
t_datos5 = np.array([3.69,3.57,3.55])/3
jexp5 = jexp(t_datos5,k_theta)
jexp5_medio = np.mean(jexp5)

for i in jexp5:
  print(f'El momento de inercia del disco soporte + cilindro hueco es: {i:.7f} kgm2')
print('-----------------------------------------------------------')

print(f'El momento de inercia medio del disco soporte + cilindro hueco es: {jexp5_medio:.7f} kgm2')

print('-----------------------------------------------------------')

#COMPARATIVA MOMENTOS INERCIA TEORICOS CILINDROS
print('COMPARATIVA MOMENTOS INERCIA TEORICOS CILINDROS')

#CILINDRO MACIZO (112mm)
print(f'El momento de inercia teórico del clindro macizo (112mm) es: {jteo_clindro:.4f} kgm2')

#CILINDRO MACIZO (44.6mm)
rcm = 0.0446
mcm = 0.339
jcm =(m*rcm**2)/2
print (f'El momento de inercia teórico del cilindro macizo (44.6mm) es: {jcm:.7f} kgm2')

#CILINDRO HUECO
mch = 0.339
rech = 0.045
rich = 0.0432
jch = (mch*(rech**2+rich**2))/2
print (f'El momento de inercia teórico del cilindro hueco es: {jch:.7f} kgm2')

print('-----------------------------------------------------------')

print (f'El momento de inercia experimental del cilindro macizo (112mm) es: {jexp2_medio:.7f}')
print (f'El momento de inercia experimental del cilindro macizo (44.6mm) es: {jexp4_medio - jexp3_medio:.7f}')
print (f'El momento de inercia experimental del cilindro hueco es: {jexp5_medio - jexp3_medio:.7f}')

print('-----------------------------------------------------------')

#MOMENTO INERCIA ESFERA
print('MOMENTO INERCIA ESFERA')
t_datos6 = np.array([5.46,5.51,5.44])/3
jexp6 = jexp(t_datos6,k_theta)
for i in jexp6:
  print(f'El momento de inercia de la esfera es: {i:.7f} kgm2')

print('-----------------------------------------------------------')
jexp6_medio = np.mean(jexp6)
print(f'El momento de inercia medio de la esfera es: {jexp6_medio:.7f} kgm2')


print('-----------------------------------------------------------')

mesf = 0.97
resf = 0.0726
jesfteo = (2/5)*(mesf*resf**2)
print (f'El momento de inercia teórico de una esfera es: {jesfteo:.7f} kgm2')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del radio (r)
r_datos = np.array([0.005,0.1,0.15,0.20,0.25])**2

# Datos del momento de inercia (j)
jmasas = momentos_masa

#Datos de la masa (m)
coeficientes = np.polyfit(r_datos, jmasas, 1)
a = coeficientes[0]  # Coeficiente angular (pendiente)
b = coeficientes[1]  # Coeficiente de posición (ordenada al origen)

#Generación de puntos del ajuste
x_fit = np.linspace(min(r_datos), max(r_datos), 100)
y_fit = a * x_fit + b

# Generación del plot
plt.figure(figsize=(10, 6))

# Dibujar los datos experimentales
plt.scatter(r_datos, jmasas, color='black', label=r'Datos experimentales ($R^2$, $J_{masas}$)', marker='o')

# Dibujar la recta de ajuste
plt.plot(x_fit, y_fit, color='red', label=fr'Recta de ajuste: $ y = {a:.4f}x + {b:.4f}$', linestyle='-')

# Etiquetas y título
plt.xlabel('Radio de giro $[m^2]$')
plt.ylabel('$J_{masas}$ $[kgm^2]$')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()

print(f'La recta de ajuste es y={a:.4f}x+{b:.4f}')
```
