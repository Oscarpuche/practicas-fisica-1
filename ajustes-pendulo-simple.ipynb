```
import numpy as np
import matplotlib.pyplot as plt

# Datos de Longitud (l)
l_datos = np.sqrt(np.array([0.420, 0.330, 0.293, 0.260, 0.220, 0.190, 0.150, 0.115, 0.080, 0.042]))

# Datos del Periodo (t)
t_datos = np.array([6.27, 5.57, 5.24, 4.96, 4.52, 4.26, 3.85, 3.27, 2.82, 2.01])/5

# Coeficientes del ajuste lineal
coeficientes = np.polyfit(l_datos, t_datos, 1)
a = coeficientes[0]
b = coeficientes[1]

# Generar puntos de la recta de ajuste para la gráfica
x_fit = np.linspace(min(l_datos), max(l_datos), 100)
y_fit = a * x_fit + b

# Cálculo de la Aceleración Gravitacional
g_calculado = (2 * np.pi / a)**2

# Resultados por pantalla
print("CÁLCULO DE LA ACELERACIÓN GRAVITACIONAL (g)")
print(f"Valor de g calculado: {g_calculado:.2f} m^2")

# Generación del plot
plt.figure(figsize=(10, 6))

# Dibujar los datos experimentales
plt.scatter(l_datos, t_datos, color='black', label=r'Datos experimentales ($\sqrt{l}$, $T$)', marker='o')

# Dibujar la recta de ajuste
plt.plot(x_fit, y_fit, color='red', label=f'Recta de ajuste: $y = {a:.4f}x + {b:.4f}$', linestyle='-')

# Etiquetas
plt.xlabel('Raíz cuadrada de la longitud del péndulo $[m^{1/2}]$')
plt.ylabel('Periodo del péndulo $[s]$')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Datos de Longitud (l)
l_datos = np.array([0.420, 0.330, 0.293, 0.260, 0.220, 0.190, 0.150, 0.115, 0.080, 0.042])

# Datos del Periodo (t)
t_datos = (np.array([6.27, 5.57, 5.24, 4.96, 4.52, 4.26, 3.85, 3.27, 2.82, 2.01])/5)**2

# Coeficientes del ajuste lineal
coeficientes = np.polyfit(t_datos, l_datos, 1)
a = coeficientes[0]
b = coeficientes[1]

# Generar puntos de la recta de ajuste para la gráfica
x_fit = np.linspace(min(t_datos), max(t_datos), 100)
y_fit = a * x_fit + b

# Cálculo de la Aceleración Gravitacional (g)
g_calculado = a * 4 * np.pi**2

# Resultados por pantalla
print("CÁLCULO DE LA ACELERACIÓN GRAVITACIONAL (g)")
print(f"Valor de g calculado: {g_calculado:.2f} m/s^2")

# Generación del plot

plt.figure(figsize=(10, 6))

# Dibujar los datos experimentales
plt.scatter(t_datos, l_datos, color='black', label='Datos experimentales ($T^2$, $l$)', marker='o')

# Dibujar la recta de ajuste
plt.plot(x_fit, y_fit, color='red', label=f'Recta de ajuste: $y = {a:.4f}x{b:.4f}$', linestyle='-')

# Etiquetas
plt.xlabel('Periodo del péndulo al cuadrado $[s^2]$')
plt.ylabel('Longitud del péndulo $[m]$')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
